```python
def img_good(path):
    img=tf.io.read_file(path)
    img=tf.image.decode_jpeg(img,channels=3)
    img=tf.image.resize(img,[224,224])
    img=tf.cast(img,tf.float32)
    img=img/255
    return img
```

```python
all_img_data=tf.data.Dataset.from_tensor_slices(path_list)
AUTOTUNE = tf.data.experimental.AUTOTUNE
data_all_list1=all_img_data.map(img_good,num_parallel_calls=AUTOTUNE)
name_data=tf.data.Dataset.from_tensor_slices((name_list,color_list))
data_all_list=tf.data.Dataset.zip((data_all_list1,name_data))
data_all_list
```

```python
data=data_all_list.take(len(path_list))
sess=len(path_list)-100
train_data=data.repeat().shuffle(buffer_size=sess).batch(20)
```

```python
my_model=tf.keras.applications.MobileNetV2(input_shape=(224,224,3),
                                          include_top=False,
                                          weights='imagenet')
```

```python
my_model.trianable = False
```

```python
inputs=tf.keras.Input(shape=(224,224,3))
```

```python
x=my_model(inputs)
```

```python
x=tf.keras.layers.GlobalAveragePooling2D()(x)
```

```python
x1=tf.keras.layers.Dense(1024,activation='relu')(x)
out_name=tf.keras.layers.Dense(4,activation='softmax',name='out_name')(x1)
x2=tf.keras.layers.Dense(1024,activation='relu')(x)
out_color=tf.keras.layers.Dense(3,activation='softmax',name='out_color')(x2)
```

```python
model=tf.keras.Model(inputs=inputs,
                    outputs=[out_name,out_color])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss={'out_color':'sparse_categorical_crossentropy','out_name':'sparse_categorical_crossentropy'},metrics=['acc'])
```

```python
nums=len(path_list)//20
```

```python
model.fit(train_data,epochs=30,steps_per_epoch=nums)
```
